```python
covid_data.drop(['Sno','Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], axis=1, inplace=True)
```

```python
covid_data.head()
```

```python
covid_data["Date"] = pd.to_datetime(covid_data["Date"],format = "%Y-%m-%d")
```

```python
covid_data.head()
```

```python
#Determing the active cases -

covid_data['Active Cases'] = covid_data['Confirmed'] - covid_data['Cured'] + covid_data['Deaths']
covid_data.head()
```

```python
#statewise

statewise = pd.pivot_table(covid_data, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc=max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
#top 10 active cases states

top_10_active_cases = covid_data.groupby(by='State/UnionTerritory').max()[['Active Cases', 'Date']].sort_values(by='Active Cases', ascending=False).reset_index()
```

```python
fig = plt.figure(figsize = (16,9))
```

```python
plt.title("Top 10 states with most active cases in India", size = 15)
```

```python
ax = sns.barplot(data=top_10_active_cases.iloc[:10], y='Active Cases', x='State/UnionTerritory', linewidth=2, edgecolor='yellow')
```

```python
#top 10 active cases states

top_10_active_cases = covid_data.groupby(by='State/UnionTerritory').max()[['Active Cases', 'Date']].sort_values(by='Active Cases', ascending=False).reset_index()
fig = plt.figure(figsize = (16,9))
plt.title("Top 10 states with most active cases in India", size = 15)
# Generate a list of colors. Use exactly the number needed for the bars.
colors = sns.color_palette("husl", len(top_10_active_cases.iloc[:10]))

# Add a dummy hue column for the plot
top_10_active_cases['dummy_hue'] = range(len(top_10_active_cases))

# Plot with different colors for each bar using hue
ax = sns.barplot(data=top_10_active_cases.iloc[:10], y='Active Cases', x='State/UnionTerritory', hue='dummy_hue', linewidth=2, edgecolor='black', palette=colors, legend=False)

# Remove the dummy hue column after plotting
top_10_active_cases.drop(columns=['dummy_hue'], inplace=True)
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Top states with highest deaths

top_10_deaths = covid_data.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by='Deaths', ascending=False).reset_index()
fig = plt.figure(figsize = (18,5))
plt.title("Top 10 states with most Deaths in India", size = 25)
# Generate a list of colors. Use exactly the number needed for the bars.
colors = sns.color_palette("viridis", len(top_10_deaths.iloc[:12]))

# Add a dummy hue column for the plot
top_10_deaths['dummy_hue'] = range(len(top_10_deaths))

# Plot with different colors for each bar using hue
ax = sns.barplot(data=top_10_deaths.iloc[:12], y='Deaths', x='State/UnionTerritory', hue='dummy_hue', linewidth=2, edgecolor='black', palette=colors, legend=False)

# Remove the dummy hue column after plotting
top_10_deaths.drop(columns=['dummy_hue'], inplace=True)
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
#Growth trends

fig = plt.figure(figsize=(12, 6))

# Plot the lineplot with correct parameters
ax = sns.lineplot(
    data=covid_data[covid_data['State/UnionTerritory'].isin(['Maharastra', 'Karnataka', 'Uttar Pradesh', 'Tamil Nadu', 'Kerala'])],
    x='Date',
    y='Active Cases',
    hue='State/UnionTerritory'
)

# Set the title of the plot
ax.set_title("Top 5 Affected States in India", size=16)
```

```python
#removes rows where state is india

vaccine = vaccine_data[vaccine_data.State!= 'India' ]
vaccine
```

```python
vaccine.rename(columns = {"Updated On": "Vaccine_Date"}, inplace = True)
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace = True)
vaccine.head()
```

```python
#most vaccinated state

max_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total", ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title("Top 5 Vaccinated States in India", size = 20)
# Generate a list of colors. Use exactly the number needed for the bars.
colors = sns.color_palette("viridis", len(max_vac.iloc[:12]))

# Add a dummy hue column for the plot
max_vac['dummy_hue'] = range(len(max_vac))

# Plot with different colors for each bar using hue
x = sns.barplot(data=max_vac.iloc[:12], y=max_vac.Total, x=max_vac.index, hue='dummy_hue', linewidth=2, edgecolor='black', palette=colors, legend = False)

# Remove the dummy hue column after plotting
max_vac.drop(columns=['dummy_hue'], inplace=True)
plt.xlabel("States")
plt.ylabel("Vaccinations")
plt.show()
```
